## Step 4. Exploratory data analysis (Python)

In addition to the data retrieved in the previous tasks, you received a second file. You now have these two CSVs:

The file /datasets/project_sql_result_01.csv contains the following data:

trips_amount: the number of rides for each taxi company from November 15th to 16th, 2017.

The file /datasets/project_sql_result_04.csv contains the following data:

dropoff_location_name: Chicago neighborhoods where the races ended

average_trips: the average number of trips that ended in each neighborhood in November 2017.


Draw conclusions based on each graph and explain the results.

```python
#importing the lybraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#### 4.1 Import the files

```python
trips_amount = pd.read_csv('sql_result_01.csv')

drop_off = pd.read_csv('moved_project_sql_result_04.csv')
```

#### 4.2 Study the data they contain

```python
# trips_amount 
# verify that the data types are correct

trips_amount.info()
print('=--' * 20)
print(trips_amount.describe())
print('=--' * 20)
print(f'The dataframe shape is: {trips_amount.shape}')
```

```python
print(trips_amount.sample(5))
```

```python
# drop_off
# verify that the data types are correct

drop_off.info()
print('=--' * 20)
print(drop_off.describe())
print('=--' * 20)
print(f'The dataframe shape is: {drop_off.shape}')
```

#### 4.3 Identify the top 10 neighborhoods in terms of destinations

```python
# top 10 neighborhood most popular
top_10_neighborhood = drop_off.sort_values(by='average_trips', ascending = False).head(10)

print(top_10_neighborhood)
```

#### 4.4 Create graphs: taxi companies and number of rides, top 10 neighborhoods by number of rides where that neighborhood is the destination.

```python
# top 10 best taxi companies in number or rides
top_10_companies = trips_amount.sort_values(by='trips_amount', ascending = False).head(10)
display(top_10_companies)
```

```python
# top 10 compnaies
# creating the bar chart

bar_colors = ['blue'] + ['red'] * 9

top_10_companies.plot(kind='bar', x='company_name', y='trips_amount', color=bar_colors, legend=False)

plt.ylabel('number of rides')
plt.xlabel('taxi companies')


plt.show()
```

```python
# top 10 neighborhoods
# creating the bar chart

colors_bars = ['yellow'] + ['green'] * 9

top_10_neighborhood.plot(kind='bar', x='dropoff_location_name', y='average_trips', color=colors_bars, legend=False)

plt.ylabel('trips')
plt.xlabel('neighborhood')


plt.show()
```
